```python
from tkinter import * 
import phonenumbers
from phonenumbers import carrier,geocoder,timezone
from timezonefinder import TimezoneFinder
from geopy.geocoders import Nominatim
from datetime import datetime
import pytz

def trackNumber():
    # get number from txt field(number.get())
    try:
        phoneOBJ = phonenumbers.parse(number.get(),None)
    
        #location ex country
        country.config(text=geocoder.description_for_number(phoneOBJ,"en"))

        #company of SIM
        sim.config(text= carrier.name_for_number(phoneOBJ,"en"))

        #Timezone ex Africa/cairo
        zone.config(text= timezone.time_zones_for_number(phoneOBJ))

        #latitude and longitude
        # from geopy.geocoders library
        geolocator = Nominatim(user_agent="phone_number_locator")
        country_code = "+" + str(phoneOBJ.country_code)
        location = geolocator.geocode(country_code)
        if location:
            longitude.config(text= location.longitude)
            latitude.config(text= location.latitude)

            # time
            obj = TimezoneFinder()
            localTime = datetime.now(pytz.timezone(obj.timezone_at(lat=location.latitude,lng=location.longitude)))
            clock.config(text=localTime.strftime("%I:%M:%p"))
        else:
            longitude.config(text="Location not found.")
            clock.config(text="Location not found.")
            latitude.config(text="Location not found.")
            
    except phonenumbers.phonenumberutil.NumberParseException:
        country.config(text="Invalid number format.")
        clock.config(text="")
        latitude.config(text="")
        longitude.config(text="")
        sim.config(text="")
        zone.config(text="")
        
        
window = Tk()
window.title("Tracking Phone Number")
window.geometry("365x584+400+100")
#cannot change size of window using mouse
window.resizable(False,False)
# to put image in icon of window
iconOfWindow = PhotoImage(file="./location.png")
window.iconphoto(False,iconOfWindow)

# Create a canvas for the top side with a specific color
TOP_canvas = Canvas(window, width=365, height=360, bg="#F3EEEA")
TOP_canvas.pack(side=TOP)
# Create a canvas for the bottom side with a specific color
BOTTOM_canvas = Canvas(window, width=365, height=292, bg="#EBE3D5")
BOTTOM_canvas.pack(side=BOTTOM)

Heading = Label(window,text="Tracking Number",font=('Bodoni MT Black',20),bg="#F3EEEA",fg="#776B5D")
Heading.place(x=57,y=100)

entryData = StringVar()
number = Entry(window,textvariable=entryData,width=17,justify="center",bd=5,font=("arial",20))
number.place(x=50,y=190)

searchBtn=Button(window,text = "Search", borderwidth=0,cursor="hand2",bg="#776B5D",fg="#F3EEEA",bd=0,command=trackNumber)
searchBtn.place(x=100,y=300, width=170, height=40, anchor='nw')

country = Label(window,text="Country",bg="#EBE3D5",fg="#776B5D",font=("arial",10,'bold'))
country.place(x=70,y=400)

sim = Label(window,text="SIM",bg="#EBE3D5",fg="#776B5D",font=("arial",10,'bold'))
sim.place(x=220,y=400)

zone = Label(window,text="TimeZone",bg="#EBE3D5",fg="#776B5D",font=("arial",10,'bold'))
zone.place(x=70,y=450)

clock = Label(window,text="Phone Time",bg="#EBE3D5",fg="#776B5D",font=("arial",10,'bold'))
clock.place(x=220,y=450)

longitude = Label(window,text="Longitude",bg="#EBE3D5",fg="#776B5D",font=("arial",10,'bold'))
longitude.place(x=70,y=500)

latitude = Label(window,text="Latitude",bg="#EBE3D5",fg="#776B5D",font=("arial",10,'bold'))
latitude.place(x=220,y=500)

window.mainloop()
```
