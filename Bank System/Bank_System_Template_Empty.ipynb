# Task 1 : 
- #### task 1.1 Make Class(User) with attributes:
    - name 
    - age 
    - gender 
    - balance
- #### task 1.2 Make method to show previous data 

```python
class Bank():
    def __init__(self,name,age,gender,balance):
        self.__name=name
        self.__age=age
        self.__gender=gender
        self.__balance=balance
    def getBalance(self):
        return self.__balance
    def getName(self):
        return self.__name
    def getAge(self):
        return self.__age
    def getGender(self):
        return self.__gender
    def displayData(self):
        print(f'name is {self.__name} , age is {self.__age} , gender is {self.__gender} , balance is {self.__balance}')
```

# Task 2 : 
- Make Child Class(Bank) which inherite from Class(user) with Mehods:
    - #### task 2.1 deposite : which take parameter(amount) and add it to balance and print new balance.  
    - #### task 2.2 withdraw : which take parameter(amount) and check if user have enough money. 
    - #### task 2.3 view balance : to show current balance. 


```python
class Alahly(Bank):
    def __init__(self,name,age,gender,balance):
        Bank.__init__(self,name,age,gender,balance)
    def viewBalance(self):
        print(f'current balance is {self.getBalance()}')
    def withDraw(self,amount):
        if int(amount) < self.getBalance():
            print('congrats, This process will happen.')
        else:
            print("This process will not be completed because your balance isnot enough")
    def deposite(self,amount):
        return int(amount)+self.getBalance()
alahly = Alahly('ahmed osama',21,'Male',20000)
alahly.deposite(200)
```
